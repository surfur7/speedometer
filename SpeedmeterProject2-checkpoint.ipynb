```python
import numpy as np
import time
import cv2 as cv
import matplotlib.pyplot as plt
#video from https://www.youtube.com/watch?v=p_dYvmis0uo&ab_channel=WindWalkTravelVideos%CA%AC 

#Video input
sampleVideo='dashcamview44.mp4'
#sampleVideo='dashcamview3.mp4'

#contrast function
def contrastAdjust(input_img, brightness = 0, contrast =0):
    if brightness != 0:
        if brightness >= 0:
            shadow = brightness
            highlight =255
        else:
            shadow =0
            highlight = 255 + brightness
        alpha_b = (highlight - shadow)/255
        gamma_b = shadow

        buf = cv.addWeighted(input_img, alpha_b, input_img, 0, gamma_b)
    else:
        buf = input_img.copy()
    if contrast != 0:
        f= 131*(contrast + 127)/(127*(131-contrast))
        alpha_c = f
        gamma_c =127*(1-f)

        buf = cv.addWeighted(buf, alpha_c, buf, 0, gamma_c)

    return buf
    

def boxdetection(input_img,test_img,origianl_img):
    O_row,O_col = origianl_img.shape[:2]
    output= input_img.copy()
    inversion = 255 - output
    ret, thresh = cv.threshold(inversion, 0, 255, 1)
    #noise filtering
    dilate = cv.dilate(thresh, None)
    erode = cv.erode(dilate,None)
    cx, cy = None ,None
    contours,hierarchy = cv.findContours(erode, cv.RETR_CCOMP,cv.CHAIN_APPROX_SIMPLE)

    for i,cnt in enumerate(contours):
        if hierarchy [0,i,3] == -1 and cv.contourArea(cnt)>100 and cv.contourArea(cnt)<9000:
            x,y,w,h = cv.boundingRect(cnt)
            cv.rectangle(test_img,(x,y),(x+w,y+h), (255,0,0),5)
            m= cv.moments(cnt)
            cx,cy = m['m10']/m['m00'],m['m01']/m['m00']
            cv.circle(test_img,(int(cx),int(cy)),3,255,-1)
    if(cx != None):
        return test_img,len(contours),cx,cy,h
    h=None
    return test_img, len(contours), h
   
```

```python
#playing original video
cap=cv.VideoCapture(sampleVideo)
cv.namedWindow(sampleVideo,cv.WND_PROP_AUTOSIZE)
if (cap.isOpened()== False): 
  print("Error opening video file")
while(True):
    ret22, originalVid= cap.read()
    time.sleep(1/30)
    if (ret22 != True or cv.waitKey(1)==ord('q')):
        break
    cv.imshow(sampleVideo,originalVid)
    
cv.destroyWindow(sampleVideo)
```

```python

cv.namedWindow(sampleVideo,cv.WND_PROP_AUTOSIZE)
cv.namedWindow('masked',cv.WND_PROP_AUTOSIZE)
cv.namedWindow('highcontrast',cv.WND_PROP_AUTOSIZE)
#OriginalImage=cv.imread("dashcam.jpg")
#img =OriginalImage.copy()
#img2= OriginalImage.copy()
cap=cv.VideoCapture(sampleVideo)
if (cap.isOpened()== False): 
  print("Error opening video file")
framecounter=0
start_time = 0
end_time = 0
speed = 0
current_speed = 0
past_speed = 0
gap = 0
past_length_mark=0
cy=0
while(True):
    
    ret, img= cap.read()
    
    #end loop when video ends
    if ret != True: 
        break
    originalframe=img.copy()

    #BIRD EYE VIEW PERSPECTIVE TRANSFORM
    H_row, W_col= img.shape[:2]
    img = img[int(0.75*H_row):int((1)*H_row), int(0.5*W_col):int(W_col)] #img slicing

    src=np.float32([[0, H_row], [950, H_row], [0,0], [W_col, 0]]) #
    #dst=np.float32([[0, H_row], [1907, H_row], [0,0], [W_col, 0]])
    dst=np.float32([[0, H_row], [110, H_row], [0,0], [W_col, 0]]) #
    #dst=np.float32([[560, H_row], [720, H_row], [0,0], [W_col, 0]]) #
    M = cv.getPerspectiveTransform(src,dst) #transformation mat
    MinV = cv.getPerspectiveTransform(dst,src) #inv transform mat
    
    warped_img= cv.warpPerspective(img,M, (W_col, H_row))
    

    H_row, W_col = warped_img.shape[:2]
    warped_img= warped_img[0:H_row, 0:int(0.25*W_col)]
    sample_warped_img=warped_img.copy()
   
    warped_img=cv.medianBlur(warped_img,5)
    warped_img = contrastAdjust(warped_img, -40, 120)
  
   
    #mask out  white
    #convert to HSV
    image_copy=warped_img.copy()
    image_copycopy= cv.cvtColor(image_copy,cv.COLOR_BGR2HSV)

    #lower_yellow=(27,230,50)
    #upper_yellow=(33,255,255)

    lower_white = (0,0,150)
    upper_white = (180,30,255)#hue saturation value white threshold
    Wimg_mask=cv.inRange(image_copycopy, lower_white, upper_white)
    
    added_mask=Wimg_mask.copy()
    
   
 

    #contouring process
    
    test_img=cv.cvtColor(added_mask.copy(),cv.COLOR_GRAY2BGR)
    # SPEED CALCULATION
    try:
        dst,number_cntr,cx,cy, length_mark =boxdetection(added_mask,test_img,originalframe)
        if(start_time == 0):
            start_time = framecounter
        elif(end_time != 0):
            gap = framecounter - start_time
            end_time = 0
            start_time = 0
            speed = (12/gap*30)*3.6
            if(past_speed != 0):
                #current_speed = (past_speed + speed) /2
                #current_speed = (0.5*past_speed+0.5*speed)
                current_speed = speed
            else:
                current_speed = 0.5*speed
            past_speed = current_speed
    except:
        dst,number_cntr,length_mark = boxdetection(added_mask,test_img,originalframe)
        if(start_time != 0):
            end_time = framecounter
    
   
    
    #img translation and merging
    inverse_warp_img= cv.warpPerspective(dst,MinV, (W_col, H_row))
    dy, dx = int(0.75*H_row), int(0.5*W_col)
    mtrx=np.float32([
        [1,0,dx],
        [0,1,dy]])
    inverse_warp_img=cv.warpAffine(inverse_warp_img,mtrx, (W_col,H_row))
    #img = img[int(0.7*H_row):int((1+0.7)*H_row), int(0.5*W_col):int(W_col)]
    #inverse_warp_img=originalframe+ inverse_warp_img
    inverse_warp_img= cv.add(originalframe,inverse_warp_img)

    
    #contourdet = dst.copy()
    contourdet = inverse_warp_img.copy()
    #contourdet[dst>0.01*dst.max()]=[0,255,0]
    
    #warped_img = added_mask

    #cv.imshow('dst',img2)
    #cv.waitKey(0)
    #cv.destroyAllWindows()
    cv.putText(contourdet, "frame counter ="+str(round(gap,1))+"frames", (700,100), cv.FONT_HERSHEY_PLAIN, 1, 255, 2)
    cv.putText(contourdet, "velocity ="+str(round(current_speed,1))+"km/h", (700,150), cv.FONT_HERSHEY_PLAIN, 1, 255, 2)
    cv.putText(contourdet, "velocity ="+str(round(current_speed/1.609,1))+"mph", (700,180), cv.FONT_HERSHEY_PLAIN, 1, 255, 2)
    
    cv.imshow(sampleVideo, contourdet)
    cv.imshow('masked',sample_warped_img)
    cv.imshow('highcontrast',added_mask)
    
    #cv.imshow('dashcam',added_mask)
    

    

    if cv.waitKey(1) == ord('q') or cv.waitKey(1) == ord('Q'):
        break
    framecounter +=1

cap.release()
cv.destroyAllWindows()
```
